```python
# -*- coding: utf-8 -*-

"""
Nombre del archivo: simon.ipynb

Descripción: Este código genera y visualiza seis estados cuánticos diferentes en esferas de 
Bloch utilizando la biblioteca QuTiP.

Autor: Fernando Ares Robledo
Fecha de creación: 5/04/2023
"""

import numpy as np
import matplotlib.pyplot as plt
from qutip import Bloch, basis

# Definir los estados cuánticos
states = [
    basis(2, 0),  # |0⟩
    basis(2, 1),  # |1⟩
    (basis(2, 0) + basis(2, 1)).unit(),  # |-⟩
    (basis(2, 0) - basis(2, 1)).unit(),  # |+⟩
    (basis(2, 0) + 1j * basis(2, 1)).unit(),  # |+i⟩
    (basis(2, 0) - 1j * basis(2, 1)).unit(),  # |-i⟩
]

# Crear las esferas de Bloch
fig, axes = plt.subplots(2, 3, figsize=(15, 10), subplot_kw={"projection": "3d"})

for ax, state, label in zip(axes.flat, states, ['|0⟩', '|1⟩', '|-⟩', '|+⟩', '|+i⟩', '|-i⟩']):
    bloch = Bloch(fig=fig, axes=ax)
    bloch.add_states(state)
    bloch.make_sphere()
    ax.set_title(label, fontsize=22, pad=20, backgroundcolor='w', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round,pad=0.2'))

# Ajustar el espacio entre las esferas de Bloch
plt.subplots_adjust(wspace=0.2, hspace=0.3)
plt.show()
```
